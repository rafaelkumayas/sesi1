```python
import numpy as np
import matplotlib.pyplot as plt

def gauss_seidel(A, B, X_init, tol=0.001, max_iter=100):
    n = len(B)
    X = X_init.copy()
    for _ in range(max_iter):
        X_new = X.copy()
        for i in range(n):
            sum_ = sum(A[i][j] * X_new[j] for j in range(n) if j != i)
            X_new[i] = (B[i] - sum_) / A[i][i]
        if np.linalg.norm(np.array(X_new) - np.array(X), ord=np.inf) < tol:
            return X_new
        X = X_new
    return X

# Persamaan pertama
a1 = np.array([[2, 3, -1], [1, -1, 2], [3, 2, 1]])
b1 = np.array([10, 4, 14])
x_init1 = [5, 5, 5]
x_gs1 = gauss_seidel(a1, b1, x_init1)
x_np1 = np.linalg.solve(a1, b1)
print("Hasil Gauss-Seidel (Persamaan 1):", x_gs1)
print("Hasil NumPy (Persamaan 1):", x_np1)

# Persamaan kedua
a2 = np.array([[3, -0.1, -0.2], [0.1, 7, -0.3], [0.3, -0.2, 10]])
b2 = np.array([7.85, -19.3, 71.4])
x_init2 = [1, 2, 3]
x_gs2 = gauss_seidel(a2, b2, x_init2)
x_np2 = np.linalg.solve(a2, b2)
print("Hasil Gauss-Seidel (Persamaan 2):", x_gs2)
print("Hasil NumPy (Persamaan 2):", x_np2)

# Regresi Linear dengan Least Squares
X = np.array([5, 4, 6, 7, 5])
Y = np.array([10, 8, 12, 14, 10])
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

plt.scatter(X, Y, color='red', label='Data Points')
plt.plot(X, m*X + c, color='blue', label=f'Regression Line: y={m:.2f}x+{c:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print(f"Nilai a (slope): {m}")
print(f"Nilai b (intercept): {c}")
```
